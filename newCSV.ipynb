```python
#Obtener informaciónd el clima en la carrera
clima = Bahrain_R.weather_data

temp = clima['AirTemp'].mean()
humi = clima['Humidity'].mean()
rain = clima['Rainfall'].mean()

print(f'La temperatura promedio es {temp}')
print(f'La humedad promedio es {humi}')
print(f'La lluvia en la {rain}')
```

```python
#Agregar la información del clima al DF
df['circuit_temp'] = temp
df['circuit_humi'] = humi
df['circuit_rain'] = rain
df.head(5)
```

```python
#Cargar la carrera de clasificación
Bahrain_Q = ff1.get_session(2023, 1, 'Q')

Bahrain_Q.load()
res_Q = Bahrain_Q.results
res_Q.head(5)
```

```python
# Actualizar los valores vacios con el último tiempo registrado de la Qualis de la carrera de clasificación
q_laptime = pd.Series(res_Q['Q3'])
q_laptime = q_laptime.fillna(res_Q['Q2'])
q_laptime = q_laptime.fillna(res_Q['Q1'])

#Transformar de datetime64 a segundos
q_laptime = q_laptime.dt.total_seconds()
q_laptime
```

```python
#Añadir la posición de la Clasificatoria y el tiempo de la mejor vuelta al DF.
df['quali_pos'] = res_Q['Position']
df['quali_laptime'] = q_laptime
```
