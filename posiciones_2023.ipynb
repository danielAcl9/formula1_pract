```python
import fastf1 as ff1
import pandas as pd
```

```python
session = ff1.get_session(2023, 1, 'R')
session.load(telemetry=False, weather=False)
res = session.results

df = pd.DataFrame()

nombre = res.BroadcastName
df['Nombre'] = nombre
```

```python
#Hay un lío con esta parte, y es que cuando se reinicie el ambiente se perderá todo el df, intentaré
#guardarlo como cvs 

for i in range(1, 13):
    num = i
    session = ff1.get_session(2023, num, 'R')
    session.load(telemetry=False, weather=False)
    res = session.results

    posFin = res.Position
    df[f'Pos. {session.event.name}'] = posFin

df
```

```python
df.to_csv('posiciones_2023.csv', index=False)
```
