```python
import fastf1 as ff1
import pandas as pd
```

```python
#Iniciar el dataframe nuevo, con los nombres de pilotos del 2022
session = ff1.get_session(2022, 22, 'R')
session.load(telemetry=False, weather=False)
res = session.results

df = pd.DataFrame()

nombre = res.BroadcastName
df['Nombre'] = nombre
df
```

```python
#Añadir la información de las carreras de 2022

for i in range(1, 22):
    num = i
    session = ff1.get_session(2022, num, 'R')
    session.load(telemetry=False, weather=False)
    res = session.results

    #Añade una columna por cada carrera
    posFin = res.Position
    df[f'Pos. {session.event.name}'] = posFin

df
```

```python
#Añadir al df existente los resultados de las carreras de 2023
df2 = df

for i in range(1, 12):
    num = i
    session = ff1.get_session(2023, num, 'R')
    session.load(telemetry=False, weather=False)
    res = session.results

    #Añade una columna por cada carrera
    posFin = res.Position
    df2[f'Pos. {session.event.name} 2023'] = posFin

df2.head(5)
```

```python
#Añade la clasificación de Zandvoort de 2022
df3 = df2

session = ff1.get_session(2022, 'Zandvoort', 'Q')
session.load(telemetry=False, weather=False)
res = session.results

posFin = res.Position
df3[f'Pos. {session.event.name} 2022 Quali Zand'] = posFin

#Añade (de nuevo) la carrera de Zandvoort de 2022

session = ff1.get_session(2022, 'Zandvoort', 'R')
session.load(telemetry=False, weather=False)
res = session.results

posFin = res.Position
df3[f'Pos. {session.event.name} 2022 Race Zand'] = posFin

df3.head(5)
```

```python
df3
```

```python
# Eliminar la columna nombre
df4 = df3.drop(['Nombre'], axis=1)
df4.head(5)
```

```python
#Transponer los resultados
df5 = df4.transpose()
df5.head(5)
```

```python
#Halla el promedio de posición de los resultados
df6 = df5.mean(axis=0, skipna=False)
df6 
```

```python
#Organizar los valores
df6.sort_values(axis=0, ascending=True, na_position='first')
```

```python
df6.to_csv('promedios_22.csv', index=False)
```
